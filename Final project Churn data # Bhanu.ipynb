```python
normalizeddata=df.iloc[:,0:17]
normalizeddata
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(normalizeddata)
normalizeddata = scaler.transform(normalizeddata)
normalizeddata=pd.DataFrame(normalizeddata)
normalizeddata
```

```python
data=df.iloc[:,-1]
data
```

```python
newdata=pd.concat([normalizeddata,data],axis=1)
```

```python
newdata
```

```python
newdata.isna().sum()
```

```python
newdata=newdata.dropna()
```

```python
newdata
```

```python
newdata.isna().sum()
```

```python
x = newdata.iloc[:,:-1]
# Target variable
y = newdata.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 0)
```

# Model building(KNN)

```python
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
#train a knn classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
#predict the target on testing set
y_pred = knn.predict(x_test)
```

```python
pd.DataFrame(y_pred)
```

```python
# Evaluate the model performance using a confusion matrix and classification report
cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(cm)
```

```python
cr = classification_report(y_test, y_pred)
print('Classification Report:')
print(cr)
```

```python
#The classifier has an accuracy of 91%, which means that it correctly predicts the class of 98% of the samples.
```

```python
#random forest classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
rfc = RandomForestClassifier(n_estimators=400, max_depth=4,max_features=0.6,max_samples=0.7)
```

```python
# Train the model on the training data
rfc.fit(x_train, y_train)
```

```python
# Make predictions on the train data
trainpred = rfc.predict(x_train)
# Make predictions on the test data
testpred = rfc.predict(x_test)
```

```python
# Evaluate the model performance on the test data
trainaccuracy = rfc.score(x_train, y_train)
trainaccuracy
```

```python
# Evaluate the model performance on the test data
testaccuracy = rfc.score(x_test, y_test)
testaccuracy
```
